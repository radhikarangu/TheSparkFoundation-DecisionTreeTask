```python
from sklearn.preprocessing import LabelEncoder
```

```python
iris=iris.drop(["Id"],axis=1)
```

```python
select_columns=["Species"]
```

```python
le=LabelEncoder()
le.fit(iris[select_columns].values.flatten())
iris[select_columns]=iris[select_columns].apply(le.fit_transform)
```

## Decision Tree Model Building

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(criterion="entropy")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=iris.iloc[:,0:4]
y=iris.iloc[:,4]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3)
```

```python
clf.fit(X_train,Y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
clf.score(X_test,Y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,y_pred)
```

## GINI index

```python
clf1=DecisionTreeClassifier(criterion="gini")
clf1.fit(X_train,Y_train)
```

```python
clf1.score(X_test,Y_test)
```

```python
from sklearn import tree
plt.figure(num=None,figsize=(15,5),dpi=70,facecolor='skyblue',edgecolor='yellow')
tree.plot_tree(clf,filled=True)
plt.title('Decision Tree')
```
